# Feature Engineering in Data Analysis
 -----------------------------------------------------------
 Steps to look at:
 1. Percent missing values
     - Drop variables that have a very high percentage of missing values
 2. Amount of variation
     - Drop values with zero variation and those that have same values all over the data
 3. Pairwise correlarions
     - If two features are highly correlated drop one of them, keep the one that has a high correlation with your target
 4. Correlation with your target
     - Features that have a low correlation with your target should be removed
