## Dataset
### California Housing prices
The data pertains to the houses found in a given California district and some summary stats about them. These are the features;
1. Longitude -A measure of how far west a house is; a higher value is farther west
2. Lattitude- A measure of how far north a house is; a higher value is farther north
3. total_rooms- Total number of rooms within a block
4. total_bedrooms- Total number of bedrooms within a block 
5. Population- Total number of people residing within a block
6. households-  Total number of households, a group of people residing within a home unit, for a block
7. Median_income- Median income for households within a block of houses (measured in tens of thousands of US Dollars)
8. Housing median age- Median age of a house within a block; a lower number is a newer building
9. Median house value- Median house value for households within a block (measured in US Dollars)
10. Ocean proximity- Location of the house w.r.t ocean/sea
